```python
# Import the library we'll use to fetch webpages
import requests
```

```python
# ============================================================================
# PART 1: UNDERSTANDING HTTP REQUESTS
# ============================================================================
# 
# When you visit a website in your browser, here's what happens:
# 1. Your browser sends a REQUEST to a server
# 2. The server processes it and sends back a RESPONSE
# 3. Your browser displays the HTML/CSS/JavaScript

print("=" * 70)
print("PART 1: MAKING YOUR FIRST HTTP REQUEST")
print("=" * 70)

# Let's fetch a simple webpage
url = 'https://www.axpo.com/group/en/about-us/values-strategy.html'
print(f"\nFetching: {url}")

response = requests.get(url)

# The response has several important pieces:
print(f"\nStatus Code: {response.status_code}")
print("✓ 200 = Success! The server is happy and sent us data")
print("✓ 404 = Not Found - the page doesn't exist")
print("✓ 500 = Server Error - something went wrong on their end")

# The actual HTML content
print(f"\nFirst 500 characters of the HTML:\n")
print(response.text[:500])
print("\n... (the rest continues)")
```

```python
# ============================================================================
# PART 2: EXPLORING THE RESPONSE
# ============================================================================

print("\n" + "=" * 70)
print("PART 2: UNDERSTANDING THE RESPONSE")
print("=" * 70)

print(f"\nResponse Headers (metadata about the response):")
print(dict(response.headers))

print(f"\nTotal length of HTML: {len(response.text)} characters")
print(f"Encoding: {response.encoding}")
```

```python
# ============================================================================
# PART 4: HTML BASICS - READING THE HTML
# ============================================================================

print("\n" + "=" * 70)
print("PART 4: UNDERSTANDING HTML STRUCTURE")
print("=" * 70)

# Let's look at a real example of HTML
sample_html = """
<html>
  <head>
    <title>My Awesome Website</title>
  </head>
  <body>
    <div class="header">
      <h1>Welcome!</h1>
      <p>This is a paragraph</p>
    </div>
    
    <div class="content">
      <a href="https://example.com" id="link-1">Click here</a>
      <span class="price">$19.99</span>
    </div>
  </body>
</html>
"""

print("\nHere's a simple HTML example:")
print(sample_html)

print("\n📚 Let's break down HTML:")
print("   - <tag> = opens an element")
print("   - </tag> = closes an element")
print("   - class='...' = a way to group similar elements")
print("   - id='...' = a unique identifier for ONE element")
print("   - Everything inside is nested (like Russian dolls)")
```

```python
# ============================================================================
# PART 5: FINDING THINGS IN HTML
# ============================================================================

print("\n" + "=" * 70)
print("PART 5: SPOTTING PATTERNS IN HTML")
print("=" * 70)

print("\n🔍 When we scrape, we're looking for patterns:")
print("   - 'Give me all <a> tags' (links)")
print("   - 'Give me all elements with class=\"price\"' (prices)")
print("   - 'Give me the text inside the <h1>' (headings)")
print("   - 'Give me the href attribute from links' (URLs)")

print("We'll use BeautifulSoup to do this (coming Thursday!)")
```

```python
# ============================================================================
# PART 6: STATIC VS DYNAMIC CONTENT
# ============================================================================

print("\n" + "=" * 70)
print("PART 6: STATIC VS DYNAMIC CONTENT")
print("=" * 70)

print("\n✅ STATIC websites:")
print("   - HTML is complete when the server sends it")
print("   - We can scrape it easily with requests + BeautifulSoup")
print("   - Most blogs, news sites, Wikipedia, etc.")

print("\n❌ DYNAMIC websites:")
print("   - HTML is empty, JavaScript loads content after")
print("   - We see the content in browser, but requests gets blank HTML")
print("   - Facebook, Twitter, Gmail, etc.")
print("   - These need special tools (Selenium, Playwright)")

print("\n💡 For Friday's challenge: You'll get a STATIC website (easy mode!).")
```
