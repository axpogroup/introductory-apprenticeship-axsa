```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
# PART 1: FETCH ALL AK-47 SKIN NAMES FROM STEAM


print("PART 1: COLLECTING AK-47 SKIN NAMES")

all_skins = []
start = 0

while True:
    url = f'https://steamcommunity.com/market/search/render/?appid=730&category_730_Weapon%5B%5D=tag_weapon_ak47&norender=1&count=10&start={start}'
    response = requests.get(url, verify=False)
    data = response.json()
    
    results_raw = data.get('results', [])
    
    if not results_raw:
        break
    
    for item in results_raw:
        all_skins.append(item.get('name'))
    
    print(f'Collected: {len(all_skins)} skins so far...')
    start += 10
    time.sleep(1)

print(f'Total skins found: {len(all_skins)}')
```

```python
# PART 2: FETCH PRICE FOR EACH SKIN

print("PART 2: FETCHING PRICES")

results = []

for skin in all_skins:
    url = f'https://steamcommunity.com/market/priceoverview/?appid=730&currency=1&market_hash_name={requests.utils.quote(skin)}'
    response = requests.get(url, verify=False)
    data = response.json()
    
    lowest = data.get('lowest_price', 'N/A')
    median = data.get('median_price', 'N/A')
    
    print(f'{skin}: {lowest} (median: {median})')
    results.append({'skin': skin, 'lowest': lowest, 'median': median})
    
    time.sleep(1)
```

```python
# PART 3: SHOW CHEAPEST AND MOST EXPENSIVE

print("PART 3: RESULTS")

# Filter out skins with no price
valid = [r for r in results if r['lowest'] != 'N/A']

# Sort by price
valid.sort(key=lambda x: float(x['lowest'].replace('$', '').replace(',', '')))

print("\nTOP 5 CHEAPEST:")
for r in valid[:10]:
    print(f"  {r['skin']}: {r['lowest']}")

print("\nTOP 5 MOST EXPENSIVE:")
for r in valid[-10:]:
    print(f"  {r['skin']}: {r['lowest']}")

print(f"\nDone. {len(results)} skins scraped.")
```
