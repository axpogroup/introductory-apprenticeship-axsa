```python
import requests

url = 'https://steamcommunity.com/market/search/render/?appid=730&category_730_Weapon%5B%5D=tag_weapon_ak47&norender=1&count=10&start=0'
response = requests.get(url, verify=False)
data = response.json()

for item in data.get('results', []):
    name = item.get('name')
    price = item.get('sell_price_text', 'N/A')
    quantity = item.get('sell_listings', 'N/A')
    print(f'{name}: {price} | Listings: {quantity}')
```
