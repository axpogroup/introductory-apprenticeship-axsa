## Part 1: What is BeautifulSoup? 🎯

BeautifulSoup is a Python library that **parses HTML and makes it easy to navigate and search**.

Think of it like this:
- **requests:** Gets the raw HTML from a website
- **BeautifulSoup:** Reads that HTML and organizes it so you can ask questions like "Give me all the links" or "Find the price"

Let's get started! 🚀

## Part 2: Import Libraries & Fetch a Webpage

```python
# Import the libraries we need
import requests
from bs4 import BeautifulSoup

# Fetch a simple webpage
url = 'https://example.com'
response = requests.get(url, verify=False)

# Check if the request was successful
print(f'Status Code: {response.status_code}')
print(f'Content Length: {len(response.text)} characters')
```

## Part 3: Parse HTML with BeautifulSoup

Now we'll convert the raw HTML into a BeautifulSoup object that we can navigate.

```python
# Create a BeautifulSoup object
# 'html.parser' tells BeautifulSoup how to parse the content
soup = BeautifulSoup(response.text, 'html.parser')

# Let's look at the structure (prettified)
print(soup.prettify()[:500])  # Print first 500 characters to see structure
```

## Part 4: Finding Elements 🔍

BeautifulSoup gives us two main ways to find elements:
1. **find()** - Returns the FIRST matching element
2. **find_all()** - Returns ALL matching elements (as a list)

Let's learn by example:

```python
# Example 1: Find the first heading (h1)
heading = soup.find('h1')
print('First H1 heading:')
print(heading)
print()

# Get just the text
if heading:
    print('Text content:', heading.get_text())
else:
    print('No H1 found')
```

```python
# Example 2: Find ALL paragraphs
paragraphs = soup.find_all('p')
print(f'Found {len(paragraphs)} paragraphs')
print()

# Loop through and print each one
for i, para in enumerate(paragraphs, 1):
    print(f'Paragraph {i}: {para.get_text()[:100]}...')  # First 100 chars
```

## Part 5: Finding by Class & ID 🎨

Classes and IDs are super useful for targeting specific elements. Use the `class_` and `id` parameters.

```python
# Find elements with a specific class
# Note: We use 'class_' (with underscore) because 'class' is a Python keyword
items = soup.find_all('div', class_='example')
print(f'Found {len(items)} divs with class="example"')
print()

# Find element with a specific ID
main = soup.find('div', id='main')
if main:
    print('Found main div!')
    print(main.get_text()[:200])
else:
    print('No element with id="main" found')
```

## Part 6: Extracting Attributes 📋

Elements can have attributes like `href` (links), `src` (images), `class`, `id`, etc.

Use `.get()` or `.attrs` to extract them.

```python
# Find all links
links = soup.find_all('a')
print(f'Found {len(links)} links')
print()

# Extract the href attribute from each link
for i, link in enumerate(links[:5], 1):  # Show first 5
    href = link.get('href')  # Get the 'href' attribute
    text = link.get_text()  # Get the link text
    print(f'Link {i}:')
    print(f'  Text: {text}')
    print(f'  URL: {href}')
    print()
```

## Part 7: CSS Selectors (Advanced) 🚀

BeautifulSoup supports CSS selectors via `.select()` - this is super powerful!

```python
# CSS selectors work just like in CSS:
# .classname = elements with this class
# #idname = element with this ID
# tag.class = tags with a specific class
# tag > child = direct children

# Example: Find all divs with class 'content'
content_divs = soup.select('div.content')
print(f'Found {len(content_divs)} divs with class="content"')

# Example: Find all paragraphs inside divs with class 'section'
section_paras = soup.select('div.section p')
print(f'Found {len(section_paras)} paragraphs inside section divs')
```

## Part 8: Real-World Example 🌍

Let's scrape actual data from a real website. We'll extract information about books.

**Note:** We're using a practice website designed for learning web scraping!

```python
# Fetch from a practice website
practice_url = 'http://books.toscrape.com/'
practice_response = requests.get(practice_url)
practice_soup = BeautifulSoup(practice_response.text, 'html.parser')

# Find all book containers
books = practice_soup.find_all('article', class_='product_pod')
print(f'Found {len(books)} books on this page')
print()

# Extract data from the first 3 books
for i, book in enumerate(books[:3], 1):
    title = book.find('h3').find('a').get('title')
    price = book.find('p', class_='price_color').get_text()
    availability = book.find('p', class_='instock availability').get_text().strip()

    print(f'Book {i}:')
    print(f'  Title: {title}')
    print(f'  Price: {price}')
    print(f'  Availability: {availability}')
    print()
```

## Part 9: Handling Errors & Debugging 🐛

Things don't always work perfectly. Here are some strategies:

```python
# Strategy 1: Check if an element exists before accessing it
element = soup.find('h1')

if element:
    print('Element found:', element.get_text())
else:
    print('Element not found')

print()

# Strategy 2: Use .get() with a default value
link = soup.find('a')
href = link.get('href', 'No href found')  # Returns default if attribute missing
print('Href:', href)

print()

# Strategy 3: Use try/except for safety
try:
    price = soup.find('span', class_='price').get_text()
    print('Price:', price)
except AttributeError:
    print('Could not find price element')
```

## Part 11: Key Takeaways 📚

**What you learned today:**

| Method | Purpose | Example |
|--------|---------|----------|
| `find()` | Find FIRST element | `soup.find('h1')` |
| `find_all()` | Find ALL elements | `soup.find_all('p')` |
| `.get_text()` | Extract text | `element.get_text()` |
| `.get()` | Extract attribute | `link.get('href')` |
| `.select()` | CSS selector | `soup.select('div.class')` |
| `.attrs` | All attributes | `element.attrs` |

**Best Practices:**
- ✅ Always check if an element exists before accessing it
- ✅ Use classes and IDs when available - they're more reliable
- ✅ Use CSS selectors for complex searches
- ✅ Test your selectors in the browser's Inspector first
- ✅ Handle errors gracefully (if element not found, don't crash)
